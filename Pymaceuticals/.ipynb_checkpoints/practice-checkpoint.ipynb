```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#mouse_metadata
#study_results
# Combine the data into a single DataFrame
mouse_study_result = pd.merge(study_results,mouse_metadata,  how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
mouse_study_result
```

```python
#find duplicate values 
duplicate_rows = pd.DataFrame(mouse_study_result[mouse_study_result.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)])
duplicate_rows['Mouse ID'].unique()
```

```python
#remove the duplicate value "g989" and create a new data set
clean_dataset=mouse_study_result.loc[mouse_study_result['Mouse ID'] !='g989']
clean_dataset
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
lasttimepoint=pd.DataFrame(clean_dataset.groupby("Mouse ID")["Timepoint"].max())
lasttimepoint_data= pd.merge(lasttimepoint,clean_dataset,  how="left", on=["Mouse ID","Mouse ID","Timepoint","Timepoint"])
#lasttimepoint_data.loc[lasttimepoint_data["Mouse ID"]=="b447"]
#clean_dataset.loc[clean_dataset["Mouse ID"]=="b447"]
#lasttimepoint_data.loc[(lasttimepoint_data["Mouse ID"]=="b447"& (lasttimepoint_data["Timepoint"]==45)) ]
lasttimepoint_data

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatmentslist=["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
treatmentanlysis=[]
outliers_list= []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in treatmentslist:
    # Locate the rows which contain mice on each drug and get the tumor volumes

    
    # add subset 
    subset_by_drug=lasttimepoint_data.loc[lasttimepoint_data["Drug Regimen"]==drug,: ]
    
    # Determine outliers using upper and lower bounds
    quartiles = subset_by_drug["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
  
    treatmentanlysis.append({"treatment":drug,                              
                                "lowerq":lowerq,
                                "upperq":upperq,
                                "iqr":iqr,
                                "lower_bound":lower_bound,
                                "upper_bound":upper_bound                            
                            })
    
    temp_outliers=[]
    outliers=pd.DataFrame(subset_by_drug.loc[(subset_by_drug["Tumor Volume (mm3)"]>upper_bound) | (lasttimepoint_data["Tumor Volume (mm3)"]<lower_bound) ,:])
    num_records = outliers.shape[0]
    print(num_records)
    if num_records!=0:        
      for index, row in outliers.iterrows():
        temp_outliers.append(row(["Mouse ID"]))
       
        

    outliers_list.append({"treatment":drug,
                          "outlierlist":[temp_outliers]
            
                        })
    
    #outliers
    #print(f"The lower quartile of temperatures is: {lowerq}")
    #print(f"The upper quartile of temperatures is: {upperq}")
    #print(f"The interquartile range of temperatures is: {iqr}")
    #print(f"The the median of temperatures is: {quartiles[0.5]} ")  
    #print(f"Values below {lower_bound} could be outliers.")
    #print(f"Values above {upper_bound} could be outliers.")


#treatmentanlysis
#treatmentanlysis=pd.DataFrame(treatmentanlysis)
#treatmentanlysis
#outliers_list
```
