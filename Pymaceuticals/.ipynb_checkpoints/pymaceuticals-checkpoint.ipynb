```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import sem
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#mouse_metadata
#study_results
# Combine the data into a single DataFrame
mouse_study_result = pd.merge(study_results,mouse_metadata,  how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
mouse_study_result
```

```python
# Checking the number of mice.
len(mouse_study_result["Mouse ID"].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = pd.DataFrame(mouse_study_result[mouse_study_result.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)])
duplicate_mice['Mouse ID'].unique()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_dataset = mouse_study_result.loc[mouse_study_result["Mouse ID"] != "g989", :]
clean_dataset
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_dataset["Mouse ID"].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
grouped_regimen = clean_dataset.groupby('Drug Regimen')['Tumor Volume (mm3)']

# mean, median, variance, standard deviation, and SEM of the tumor volume. 

MeanTumorVolume=grouped_regimen.mean()
MedianTumorVolume=grouped_regimen.median()
TumorVolumeVariance=grouped_regimen.var()
TumorVolumeStdDev=grouped_regimen.std()
TumorVolumeStdErr=grouped_regimen.sem()

# Assemble the resulting series into a single summary DataFrame.
Drug_Regimen_summary=pd.DataFrame({
                                   "Mean Tumor Volume":MeanTumorVolume,
                                   "Median Tumor Volume":MedianTumorVolume,
                                   "Tumor Volume Variance":TumorVolumeVariance,
                                   "Tumor Volume StdDev":TumorVolumeStdDev,
                                   "Tumor Volume StdErr":TumorVolumeStdErr                                   
    
                                  })
Drug_Regimen_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
clean_dataset.groupby('Drug Regimen').agg({"Tumor Volume (mm3)":['mean', 'median', 'var', 'std','sem']})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
lasttimepoint=pd.DataFrame(clean_dataset.groupby("Mouse ID")["Timepoint"].max())
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
lasttimepoint_data= pd.merge(lasttimepoint,clean_dataset,  how="left", on=["Mouse ID","Mouse ID","Timepoint","Timepoint"])
lasttimepoint_data
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatmentslist=["Capomulin","Ramicane","Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
treatmentanlysis=[]
outliers_list= []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for drug in treatmentslist:
    # Locate the rows which contain mice on each drug and get the tumor volumes
  
    # add subset 
    subset_by_drug=lasttimepoint_data.loc[lasttimepoint_data["Drug Regimen"]==drug,: ]
    
    # Determine outliers using upper and lower bounds
    quartiles = subset_by_drug["Tumor Volume (mm3)"].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
  
    treatmentanlysis.append({"treatment":drug,                              
                                "lowerq":lowerq,
                                "upperq":upperq,
                                "iqr":iqr,
                                "lower_bound":lower_bound,
                                "upper_bound":upper_bound                            
                            })
     # Determine outliers using upper and lower bounds
    temp_outliers=[]
    outliers=pd.DataFrame(subset_by_drug.loc[(subset_by_drug["Tumor Volume (mm3)"]>upper_bound) | (lasttimepoint_data["Tumor Volume (mm3)"]<lower_bound) ,:])
    num_records = outliers.shape[0]
 
    if num_records!=0:        
        for outliers["Mouse ID"], row in outliers.iterrows():
            temp_outliers.append(outliers["Mouse ID"])
                  
    outliers_list.append({"treatment":drug,
                          "outlierlist":[temp_outliers]
            
                        })
#List outlires 
outliers_list=pd.DataFrame(outliers_list)
for index, row in outliers_list.iterrows():
    print(f"{row['treatment']},'s potential outliers (Mouse ID's) :  {row['outlierlist']}")
```

```python
#Print statistical summary
treatmentanlysis=pd.DataFrame(treatmentanlysis)
treatmentanlysis
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
selectedregims=clean_dataset[clean_dataset["Drug Regimen"].isin(treatmentslist)]
drugboxplot=selectedregims.boxplot(column="Tumor Volume (mm3)",by="Drug Regimen",figsize=(8,4))
drugboxplot.set_ylabel("Final Tumor Volume (mm3)")
drugboxplot.set_title("")
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
singlemouseb128=clean_dataset[(clean_dataset["Mouse ID"]=="b128") &(clean_dataset["Drug Regimen"]=="Capomulin")]

# Create the line chart
plt.plot(singlemouseb128["Timepoint"], singlemouseb128["Tumor Volume (mm3)"])

# Add labels and title
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin Treatment of mouse b128")

# Display the plot
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_data=clean_dataset.loc[clean_dataset["Drug Regimen"]=="Capomulin"]
capomulingrp_data=capomulin_data.groupby("Weight (g)")["Tumor Volume (mm3)"].mean().reset_index()
x_values=capomulingrp_data["Weight (g)"]
y_values=capomulingrp_data["Tumor Volume (mm3)"]
plt.scatter(x_values, y_values, marker="o", facecolors="red", edgecolors="black",
            s=x_values, alpha=0.75)
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor vulume(mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
#correlation coefficient
print(f"The correlation coefficient between mouse weight and average observed tumor volume is {round(st.pearsonr(x_values,y_values)[0],2)}")

#linear regression model
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,38),fontsize=15,color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor vulume(mm3)")
print(f"The r-squared is: {rvalue**2}")
plt.show()

```
